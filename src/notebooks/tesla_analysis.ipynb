```python
from tesla_scraper import TeslaScraper
import pandas as pd
```

```python
scraper = TeslaScraper()
data = scraper.parse()
```

```python
df = pd.DataFrame(data)
df
```

```python
df_sorted = df.sort_values(by=["price", "miles"])
df_sorted
```
